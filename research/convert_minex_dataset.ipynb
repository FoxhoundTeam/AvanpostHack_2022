Преобразует датасет https://github.com/usnistgov/minex/tree/master/minexiii/validation/validation_imagery_raw из бинарного "сырого" формата в PNG

# Получение имён файлов и разрешений из header-файла

```python
# массив имён и разрешений

# "minex/minexiii/validation/minexiii_validation.h"

header_file = open("minex/minexiii/validation/minexiii_validation_data.h", "r")
header_file_contents = header_file.read()
header_file.close()
header_file_contents[:1000]
```

```python
import re
reg_exp = r"(\s*{\s*\"(\w+.gray)\",\s*{(\d{3}),\s*(\d{3}),\s*(\w+),\s*(\w+),\s*(\w+)}\s*},*)"
# search_results = re.findall(reg_exp, header_file_contents)
# search_results
dataset = []
for item in re.finditer(reg_exp, header_file_contents):
    filename = item.group(2)
    width = item.group(3)
    height = item.group(4)
    quality = item.group(5)
    fingername = item.group(6)
    origin = item.group(7)
    dataset.append({"filename":filename, "width":int(width), "height":int(height)})
dataset[:10]
```

# Сохранение файлов в формате PNG

```python
import cv2
import numpy as np

for item in dataset:
    filename = item["filename"]
    width = item["width"]
    height = item["height"]
    image_np = np.fromfile(f"minex/minexiii/validation/validation_imagery_raw/{filename}" ,dtype=np.uint8)
    image_np = image_np.reshape(( height, width));
    # image_np.view()
    cv2.imwrite(f"dataset\\{filename}.png", image_np)
```
